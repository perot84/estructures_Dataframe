```python
page = requests.get('http://quotes.toscrape.com')

# Parsear el HTML de la página
soup = BeautifulSoup(page.content, 'html.parser')

# <span class="text" itemprop="text">
textos_b= soup.find_all('span', class_='text')
print ('Textos')
for texto_b in textos_b:
    print(texto_b.text)

#<span>by <small class="author" itemprop="author">
autores_b= soup.find_all('small', class_='author')
print ('Autores')
for autor_b in autores_b:
    print (autor_b.text)
#<a class="tag" href="/tag/deep-thoughts/page/1/">deep-thoughts</a>    
print ('Tags')    
tags= soup.find_all('a', class_='tag')
for tag in tags:
    print(tag.text)
```

Webscraping amb Scrapy de www.lacentral.com

Codi de l'spider:


import scrapy
from html import unescape

class CentralSpider(scrapy.Spider):
    name = 'central'
    start_urls = ['https://www.lacentral.com/recomendados/?mat=S']

    def parse(self, response):
        authors = response.css('h4 a')
        for author in authors:
            yield {
                'author': unescape (author.css('::text').get().strip())
                }
        titles = response.css('h5 a')
        for title in titles:
            yield {
                'title': unescape (title.css('::text').get().strip())
                }
